```python
greenmark['greenmarkrating'] = greenmark['greenmarkrating'].fillna("Uncertified")
greenmark['greenmarkyearofaward'] = greenmark['greenmarkyearofaward'].fillna(0)
```

```python
def retrievepostalcode(address): #Retrieving postal code from address
    return int(address[-6:])

def greenmarkscore(cert): #Dummy data creation
    if cert == 'Uncertified':
        return random.randint(1, 40)
    elif cert == 'Certified' or cert == 'Legislated':
        return random.randint(41, 60)
    elif cert == 'Gold':
        return random.randint(61, 70)
    elif cert == 'GoldPlus':
        return random.randint(71, 80)
    else:
        return random.randint(81, 100)
        
```

```python
greenmark['postalcode'] = greenmark['buildingaddress'].apply(retrievepostalcode)
greenmark['greenmarkscore'] = greenmark['greenmarkrating'].apply(greenmarkscore)
```

```python
meanCarbonEmissions = greenmark.iloc[:,-5:-2].mean(axis = 1)
```

```python
buildings = greenmark.filter(['postalcode', 'buildingaddress', 'buildingname', 'mainbuildingfunction'])
buildings['latitude'] = ""
buildings['longitude'] = ""
```

```python
building_details = greenmark.filter(['postalcode', 'greenmarkscore'])
building_details['notable_qualities'] = ""
building_details['other_awards'] = ""
building_details['carbon_emissions'] = greenmark.iloc[:,-5:-2].mean(axis = 1).round(3)
```
